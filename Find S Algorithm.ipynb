```python
def find_s_algorithm(data):
    # Initialize the hypothesis as the first positive example
    hypothesis = None
    for example in data:
        if example[-1].lower() == 'yes':
            hypothesis = example[:-1]
            break

    if hypothesis is None:
        return "No positive examples in the dataset."

    for example in data:
        if example[-1].lower() == 'yes':
            for i in range(len(hypothesis)):
                if hypothesis[i] != example[i]:
                    hypothesis[i] = '?'

    return hypothesis


def main():
    n = int(input("Enter number of training examples: "))
    data = []

    print("Enter each example's attributes followed by label (Yes/No), separated by space.")
    for _ in range(n):
        example = input().strip().split()
        data.append(example)

    hypothesis = find_s_algorithm(data)
    print("Learned Hypothesis:", hypothesis)


if __name__ == "__main__":
    main()
```
