```python
import pandas as pd


data = {
    'age': ['senior', 'young', 'young', 'middle', 'young', 'senior'],
    'income': ['low', 'high', 'medium', 'high', 'low', 'medium'],
    'student': ['no', 'no', 'yes', 'yes', 'yes', 'no'],
    'credit_rating': ['fair', 'fair', 'excellent', 'excellent', 'fair', 'fair'],
    'employment': ['unemployeed', 'employeed', 'employeed', 'employeed', 'unemployeed', 'employeed'],
    'collateral': ['no', 'yes', 'no', 'yes', 'no', 'yes'],
    'loan_approval': ['yes', 'no', 'no', 'yes', 'yes', 'yes']
}


df = pd.DataFrame(data)

S = ['?' for _ in df.columns[:-1]]
G = [['?' for _ in df.columns[:-1]]]

def satisfies_instance(hypothesis, instance):
    for h, i in zip(hypothesis, instance):
        if h != '?' and h != i:
            return False
    return True

def candidate_elimination_algorithm(df):
    global S, G
    for i, row in df.iterrows():
        instance = row[:-1].values.tolist()
        target = row['loan_approval']

        if target == 'yes':
            G = [g for g in G if satisfies_instance(g, instance)]
            for j in range(len(S)):
                if not satisfies_instance(S[j], instance):
                    S[j] = ['?' if h != i else h for h, i in zip(S[j], instance)]

        elif target == 'no':
            for j in range(len(G)):
                if satisfies_instance(G[j], instance):
                    G[j] = ['?' if h != i else h for h, i in zip(G[j], instance)]

    return S, G

S, G = candidate_elimination_algorithm(df)

print("Specific Hypothesis (S):", S)
print("General Hypothesis (G):", G)

```
